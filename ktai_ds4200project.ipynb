```python
# Kimberly Tai - DS4200 Final Project Code
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
```

```python
ev = pd.read_csv("Electric_Vehicle_Population_Data.csv")
ev.head()
```

```python
makes = ev.groupby(['Make', 'Model']).size()
makes = pd.DataFrame(makes).reset_index()
```

```python
makes.columns = ['make', 'model', 'count']
makes
```

```python
make_list = list(makes.make.unique())
```

```python
# A dropdown filter
make_dropdown = alt.binding_select(options=make_list, name="Make: ")
make_select = alt.selection_point(fields=['make'], bind=make_dropdown)
```

```python
base = alt.Chart(makes).mark_bar().encode(
    x=alt.X('model:N'),
    y=alt.Y('count:Q'),
    tooltip=['make:N', 'model:N', 'count:Q']
).properties(
    width=900
)
```

```python
filter_make = base.add_params(
    make_select
).transform_filter(
    make_select
).properties(title="Make and Model Counts")
```

```python
top_ten = pd.DataFrame(ev.groupby(["Make"]).size()).reset_index()
```

```python
top_ten.sort_values(0, inplace=True, ascending=False)
top_ten.head(10)
```

```python
top_ten_makes = top_ten.head(10).Make
list(top_ten_makes)
```

```python
years = ev.groupby(['Model Year', 'Make']).size()
years = pd.DataFrame(years).reset_index()
years.columns = ['year', 'make', 'count']

years.head()
```

```python
top_ten_years = years[years['make'].isin(list(top_ten_makes))]
top_ten_years
```

```python
base = alt.Chart(years).mark_bar().encode(
    x=alt.X("make:N"),
    y=alt.Y("count:Q"),
    tooltip=["make", "count"]
).properties(title="EVs per Model Year")
```

```python
# A slider filter
year_slider = alt.binding_range(min=1999, max=2025, step=1, name="Year: ")
slider_selection = alt.selection_point(bind=year_slider, fields=['year'])

filter_year = base.add_params(
    slider_selection
).transform_filter(
    slider_selection
)
```
